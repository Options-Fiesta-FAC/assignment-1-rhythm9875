```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

Taking inputs

```python
K = float(input("Enter the value of Strike price: "))
P = float(input("Enter Premium : "))
m_range = float(input("Enter Minimum Spot Price (e.g. K - 50): "))
M_range = float(input("Enter Maximum Spot Price (e.g. K + 50): "))
```

```python
S = np.linspace(K - 50, K + 50, 200)
```

Defining Payoff Function

```python
def buy_call_payoff(S, K , P):
    return np.maximum(S - K, 0) - P

def sell_call_payoff(S, K, P):
    return P - np.maximum(S - K, 0)

def buy_put_payoff(S, K, P):
    return np.maximum(K - S, 0) - P

def sell_put_payoff(S, K, P):
    return P - np.maximum(K - S, 0)
```

```python
y1 = buy_call_payoff(S, K, P)
y2 = sell_call_payoff(S, K, P)
y3 = buy_put_payoff(S, K, P)
y4 = sell_put_payoff(S, K, P)
```

```python
# Buy_call
plt.subplot(2,2,1)
plt.plot(S, y1)
plt.title('Buy Call')
plt.xlabel('Spot Price')
plt.ylabel('Payoff')

# Sell_call
plt.subplot(2,2,2)
plt.plot(S, y2)
plt.title('Sell Call')
plt.xlabel('Spot Price')
plt.ylabel('Payoff')

# Buy_put
plt.subplot(2,2,3)
plt.plot(S, y3)
plt.title('Buy Put')
plt.xlabel('Spot Price')
plt.ylabel('Payoff')

# Sell_put
plt.subplot(2,2,4)
plt.plot(S, y4)
plt.title('Sell Put')
plt.xlabel('Spot Price')
plt.ylabel('Payoff')

plt.tight_layout()
plt.show()
```
